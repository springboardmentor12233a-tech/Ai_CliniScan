```python
output_folder = "/kaggle/working/png_images"

if os.path.exists(output_folder):
    shutil.rmtree(output_folder)

os.makedirs(output_folder)

print("Fresh folder created ✅")
```

```python
train_folder = "/kaggle/input/vinbigdata-chest-xray-abnormalities-detection/train"
test_folder = "/kaggle/input/vinbigdata-chest-xray-abnormalities-detection/test"
output_folder = "/kaggle/working/png_images"

# Get all dicom files from train and test
train_files = [os.path.join(train_folder, f) for f in os.listdir(train_folder) if f.endswith(".dicom")]
test_files = [os.path.join(test_folder, f) for f in os.listdir(test_folder) if f.endswith(".dicom")]

all_files = train_files + test_files

print("Total images found:", len(all_files))

# If you want only first 7000 images:
limit = 7000
all_files = all_files[:limit]

print("Converting", len(all_files), "images...")

for file_path in tqdm(all_files):
    try:
        ds = pydicom.dcmread(file_path)
        image = ds.pixel_array

        # Normalize
        image = (image - np.min(image)) / (np.max(image) - np.min(image))
        image = (image * 255).astype(np.uint8)

        # Resize to 1024x1024
        image = cv2.resize(image, (1024, 1024))

        # Save
        filename = os.path.basename(file_path).replace(".dicom", ".png")
        output_path = os.path.join(output_folder, filename)

        cv2.imwrite(output_path, image)

    except:
        print("Error in:", file_path)

print("✅ Conversion completed!")
```
